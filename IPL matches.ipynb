# bar plot: To understand the distribution of categorical  cloumn

```python
keys_top5=list(data['player_of_match'].value_counts()[0:5].keys())
values_top5 =list(data['player_of_match'].value_counts()[0:5])
plt.figure(figsize=(8,5))
plt.bar(keys_top5,values_top5, color='green')
plt.xlabel('Name of the player')
plt.ylabel('Number of man of the match awards ')
plt.title('Top 5 players ')
```

```python
# Extracting a reconds where team won batting first
batting_first= data[data['win_by_runs']!=0]
batting_first['winner'].value_counts()
```

```python
data.head(2)
```

```python
# making an histogram(numberical data)to find out in which  range of runs  batting first team won
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
#Find out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
# Bar plot for the top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),batting_first['winner'].value_counts()[0:3],color=['blue','Yellow','orange'])
plt.show()
```

```python
# which team has the highest percentage after batting first
# making a pie chart 
# for the pie chat we will give numerical values first
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
